```python
import pandas as pd
import numpy as np
from ast import literal_eval
from konlpy.tag import Okt
```

```python
import pandas as pd
import numpy as np

# pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 4)
pd.set_option('display.width', 1000)

netflix_data = pd.read_csv('/content/final_netflix.csv', encoding = 'cp949')
netflix_data['Summary'] = netflix_data['Summary'].str.replace("[^\w]"," ")

netflix_data.head(1)
```

```python
##TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vec = TfidfVectorizer(ngram_range=(1,2))

# fit ->  평균과 표준편차 계산 , transform -> 정규화
stopwords = ['은','는','이','가','을','를','에','에게', '으로', '다', '의', '로', '까지', '열',
             '에서', '하다', '흉', '뒤', '뿐', '이다', '과', '들', '안', '밖', '에서는', '혈', '여',
             '그', '한', '살', '게', '같다', '네', '뭐', '씌었', '내', '하', '랴', '들', '건']

X_train = []
for stc in netflix_data['Summary']:
    token = []
    words = Okt().morphs(stc, stem=True)
    for word in words:
        if word not in stopwords:
            token.append(word.lower())
    X_train.append(token)
```

```python
X1_train = []
for stc in X_train:
  Summary_data = " ".join(stc)
  X1_train.append(Summary_data)
X2_train = pd.Series(X1_train)

tfidf_matrix = tfidf_vec.fit_transform(X2_train)

print(tfidf_vec.vocabulary_.items())
print(tfidf_matrix)
```

```python
#유사도 행렬 만들기(tfidf 특성)
from sklearn.metrics.pairwise import cosine_similarity

#자신과의 유사도는 1
# 두 벡터의 내적 cos(tfidf_matrix)
genres_similarity = cosine_similarity(tfidf_matrix,tfidf_matrix)
print(genres_similarity)

#유사도 값이 높은 영화의 제목
#유사도 값이 높은 순으로 인덱스 값을 뽑아낸다.
similar_index = np.argsort(-genres_similarity)
print(similar_index)
```

```python
#################################
   # 1차 시도의 문제점 발견. #
#################################
## 줄거리만 학습하다보니 특징이 비슷하지 않아도 줄거리에서 나온 단어가 비슷하면, 결과를 출력시키게 됨

# input_movie = input()

# netflix_data_index = netflix_data[netflix_data['Title'] == input_movie].index.values
# #유사도 기반으로 20개까지 추린다.
# similar_netflix_data = similar_index[netflix_data_index,:20]

# #유사한 영화 인덱스 출력
# # 2차원 벡터를 1차원으로
# similar_netflix_data_index =  similar_netflix_data.reshape(-1)
# netflix_d_data = netflix_data.iloc[similar_netflix_data_index]

# print(netflix_d_data)
```

```python
input_movie = input()

netflix_data_index = netflix_data[netflix_data['Title'] == input_movie].index.values
#유사도 기반으로 순서대로 나열한다.
similar_netflix_data = similar_index[netflix_data_index,:2100]
#유사한 영화 인덱스 출력
# 2차원 벡터를 1차원으로
similar_netflix_data_index =  similar_netflix_data.reshape(-1)
netflix_d_data = netflix_data.iloc[similar_netflix_data_index]

count = 0
Final_Feature_series = []
for index in netflix_d_data.index:
  dic_total_data = []
  if count ==0:
    Feature_target = netflix_d_data.loc[index,'Feature'].split()
    target_dict = {'title':netflix_d_data.loc[index,'Title'],
                   'genre':netflix_d_data.loc[index,'Genre'],
                   'Feature':Feature_target}
    count+=1
    
  else:
    Feature_series = []
    Feature_series.append(netflix_d_data.loc[index,'Feature'])
    Feature_dic =  {'title':netflix_d_data.loc[index,'Title'],
                    'genre':netflix_d_data.loc[index,'Genre'],
                    'Feature':Feature_series}
    for features in Feature_dic['Feature']:
      Feature_dic['Feature'] = features.split(" ")
    Final_Feature_series.append(Feature_dic)
print(Final_Feature_series)
```

```python
count = 0
word = []
for stc in Final_Feature_series:
  sum_list = stc['Feature'] + target_dict['Feature']
  my_list = set(sum_list)
  final_list = list(my_list)
  if len(final_list) == len(stc['Feature'] + target_dict['Feature']):
    continue
  else:
    if len(final_list)/2  <= len(target_dict['Feature']):
      if count == 15:
        break
      else:
        print(stc['title'])
        word.append(stc['genre'])

        count+=1
```

```python
word_list = []
for i in word:
  for j in i.split():
    word_list.append(j)

print(word_list)
```

```python
sw_words = ['및', 'TV', '영화', '프로그램', '원작의', '바탕', '만화를']
clean_word_list = []

for i in word_list:
  if i not in sw_words:
    clean_word_list.append(i)
```

```python
from collections import Counter

count_list = Counter(clean_word_list)
print(count_list)
```

```python
common_cl = count_list.most_common(10)
print(common_cl)
```

```python
common_cl_dict = {}

#빈도순위대로 0위~9위 (빈도수가 가장 높은게 0등)
#딕셔너리 형태로, '단어':빈도순위
i = 0
for (key, value) in common_cl:
    common_cl_dict[key] = i
    i = i+1

print(common_cl_dict)
```

```python
# one-hot 인코딩

oh_vector_list = []

for value in common_cl_dict.values():
    oh_vector = [0] * len(common_cl_dict)
    oh_vector[value] = 1
    oh_vector_list.append(oh_vector)

print(oh_vector_list)
```
